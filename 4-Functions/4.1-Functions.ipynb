```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2 == 0:
        print("Evn")
    else:
        print("Odd")
even_or_odd(5)
```

```python
## function with multiple parameters
def add(a,b):
    return a+b
result = add(2,4)
print(result)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome to the paradise")

greet("Krish")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome to the paradise")

greet()
```

```python
### Variable length arguments

## Positional and keyword arguments
def print_number(*args):
    for n in args:
        print(n)

print_number(1,2,3,4,5,6,7,8)
```

```python
### Keyword Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Krish",age="32",country="India")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,5,name="Krish",age="32",country="India")
```
