Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and dataFrame. A series is a one-dimensional array-like object, while a DataFrame is a two dimensional, size-mutable, and profitablity heterogeneous tabular data structure with labeled axes (rows and coulumns).

```python
import pandas as pd
```

```python
import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
data={'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## create a Dataframe from a dictionary of list

data={
    'Name': ['Krish','John','Jack'],
    'Age': [25,30,45],
    'City': ['Bangalore','New York', 'Florida']
}

df=pd.DataFrame(data)
print(df)
```

```python
## Create a Data Frame froma list of Dictionaries
data=[
    {'Name':'Krish', 'Age':32,'City': 'Bangalore'},
    {'Name':'John', 'Age':32,'City': 'Bangalore'},
    {'Name':'Bobby', 'Age':32,'City': 'Bangalore'},
    {'Name':'Jack', 'Age':32,'City': 'Bangalore'}
]

df=pd.DataFrame(data)
print(df)
```
