```python
## while loop
count=0
while count < 5:
    print(count)
    count=count+1
```

```python
count=0
while count%2 == 0:
    print(count)
    count=count+1
    
```

```python
## loop control statement
## break - exits the loop prematurely
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass - null operation, it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples - Calculate the sum of first N natural numbers using while 

n=10
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1

print(F"Sum of 10 natural number is {sum}")
```

```python
## Examples - Calculate the sum of first N natural numbers using for loop

n=10
sum=0
for i in range(1,11):
    sum=sum+i
print(F"Sum of 10 natural number is {sum}")
```

```python
## Example - prime number between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
