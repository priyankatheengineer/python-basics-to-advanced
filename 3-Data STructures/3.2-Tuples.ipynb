```python
## Packing and Unpacking tuples - in Python, commas create a tuple â€” not parentheses.

## Python automatically convert comma-separated values into a tuple. It is a packed tuple.

packed_tuple=1,"hello",3.14
print(packed_tuple)
```

```python
## Unpacking
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1, 2, 3, 4, 5, 6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
nested_tuple = ((1,2,3),("a","b","c"),(True, False))
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
```
