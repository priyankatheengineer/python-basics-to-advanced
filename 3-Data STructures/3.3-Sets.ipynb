##### 
| Feature                | **List**                   | **Tuple**           | **Set**              | **Dictionary**                |
| ---------------------- | -------------------------- | ------------------- | -------------------- | ----------------------------- |
| **Syntax**             | `[ ]`                      | `( )`               | `{ }`                | `{key: value}`                |
| **Example**            | `[1, 2, 3]`                | `(1, 2, 3)`         | `{1, 2, 3}`          | `{"a": 1, "b": 2}`            |
| **Ordered**            | ✅ Yes                      | ✅ Yes               | ❌ No (unordered)     | ✅ Yes (Python 3.7+)           |
| **Mutable**            | ✅ Yes                      | ❌ No                | ✅ Yes                | ✅ Yes                         |
| **Duplicates Allowed** | ✅ Yes                      | ✅ Yes               | ❌ No                 | ❌ Keys: No <br> ✅ Values: Yes |
| **Indexing**           | ✅ Yes                      | ✅ Yes               | ❌ No                 | ❌ (Access via keys)           |
| **Key-Value Pair**     | ❌ No                       | ❌ No                | ❌ No                 | ✅ Yes                         |
| **Use Case**           | Store ordered collection   | Fixed/constant data | Unique elements      | Store mapped data             |
| **Performance**        | Slightly slower than tuple | Faster than list    | Fast membership test | Fast key lookup               |


```python
my_set={1,2,3,4,5}
print(type(my_set))
```

```python
my_empty_set=set()
print(my_empty_set)
```

```python
my_set=set([1,2,3,6,5,4,5,6])
print(my_set)
```

```python
## basic set operations
## Adding and removing elements
my_set.add(7)
print(my_set)
```

```python
## remove the lements
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

my_set.discard(11)

```python
## pop method
removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
my_set.clear()
print(my_set)
```

```python
## set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematical operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## union
union_set=set1.union(set2)
print(union_set)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## union
intersection_set=set1.intersection(set2)
print(intersection_set)
```

```python
set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
print(set1.difference(set2))
```

```python
## symmetric difference
set1.symmetric_difference(set2)
```

```python
## sets methods
set1={1,2,3,4,5}
set2={3,4,5}
print(set1.issubset(set2))


print(set1.issuperset(set2))
```
